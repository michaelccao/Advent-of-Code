```python
from pathlib import Path
```

```python
data = Path("../Data/Day09.txt").read_text().splitlines()

tiles = [[int(coord) for coord in box.split(",")] for box in data]
```

```python
def largest_rectangle(tiles: list[list[int]]) -> tuple[int, int]:
    p1 = 0
    p2 = 0

    horiz = set()
    vert = set()

    for i in range(len(tiles)):
        x0, y0 = tiles[i]
        x1, y1 = tiles[(i+1) % len(tiles)]

        if x0 == x1:
            vert.add((x0, min(y0, y1), max(y0, y1)))
        else:
            horiz.add((y0, min(x0, x1), max(x0, x1)))
    
    for i in range(len(tiles)):
        x0, y0 = tiles[i]

        for j in range(i+1, len(tiles)):
            x1, y1 = tiles[j]

            area = (abs(x0-x1)+1)*(abs(y0-y1)+1)

            p1 = max(p1, area)

            if area > p2:
                left = min(x0, x1)
                right = max(x0, x1)
                up = min(y0, y1)
                down = max(y0, y1)

                left_wall = (left, up, down)
                right_wall = (right, up, down)

                top_wall = (up, left, right)
                bot_wall = (down, left, right)

                

                
    
    return p1, p2


```

```python
largest_rectangle(tiles)
```
