```python
from pathlib import Path
```

```python
data = Path("../Data/Day09.txt").read_text().splitlines()

tiles = [[int(coord) for coord in box.split(",")] for box in data]
```

```python
def largest_rectangle(tiles: list[list[int]]) -> tuple[int, int]:
    p1 = 0
    p2 = 0

    edges = get_edges(tiles)
    cache = {}
    
    for i in range(len(tiles)):
        x0, y0 = tiles[i]

        for j in range(i+1, len(tiles)):
            x1, y1 = tiles[j]

            area = (abs(x0-x1)+1)*(abs(y0-y1)+1)

            p1 = max(p1, area)

            if area > p2:
                left = min(x0, x1)
                right = max(x0, x1)
                up = min(y0, y1)
                down = max(y0, y1)

                for wall in [
                    (left, left, up, down),
                    (right, right, up, down),
                    (left, right, up, up),
                    (left, right, down, down),
                ]:
                    if not wall_inside(wall, edges, cache):
                        break
                else:
                    p2 = area
    
    return p1, p2

# Given wall with unknown inside/outside: ????????????
# Check for intersection with edges

# If more than one unknown section: ##????###?????###
# Then wall cannot be inside

# If only one unknown section ##??????###
# Check one point inside unknown section


def wall_inside(wall: tuple[int, int, int, int], edges: set[int, int, int, int], cache: dict[tuple[int, int], bool] = {}) -> bool:
    x0, x1, y0, y1 = wall
    
    for x in range(x0, x1+1):
        for y in range(y0, y1+1):
            if not point_inside(x, y, edges, cache):
                print(x, y)
                return False
    
    return True

def intersection(wall: tuple[int, int, int, int], edge: tuple[int, int, int, int]) -> list[tuple[int, int, int, int]]:
    wx0, wx1, wy0, wy1 = wall
    ex0, ex1, ey0, ey1 = edge

    if wx0 == wx1: # Vertical wall
        if ex0 == ex1 == wx0: # Colinear vertical edge
            pass
        elif ey0 == ey1 and ex0 <= wx0 <= ex1: # Horizontal edge
            if ey0 == wy0:
                return [(wx0, wx0, wy0+1, wy1)]
            
def point_inside(x: int, y: int, edges: set[int, int, int, int], cache: dict[(int, int), bool] = {}) -> bool:
    if (x, y) in cache:
        return cache[x, y]
    
    # Go from x, y to x, inf
    crossings = set()

    for x0, x1, y0, y1 in edges:

        if x0 == x == x1 and y1 >= y:
            for yx in range(max(y, y0), y1+1):
                crossings.add(yx)

        elif x0 <= x <= x1 and y1 >= y:
            crossings.add(y1)

    num_crossings = 0

    for cross in crossings:
        if cross+1 not in crossings:
            num_crossings += 1


    cache[x, y] = num_crossings % 2 == 1

    return cache[x, y]

def get_edges(tiles: list[list[int]]) -> set[tuple[int, int, int, int]]:
    edges = set()

    for i in range(len(tiles)):
        x0, y0 = tiles[i]
        x1, y1 = tiles[(i+1) % len(tiles)]

        edges.add((min(x0, x1), max(x0, x1), min(y0, y1), max(y0, y1)))

    return edges


```
