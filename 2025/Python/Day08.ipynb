```python
from pathlib import Path
import networkx as nx
```

```python
data = Path("../Data/Day08.txt").read_text().splitlines()

boxes = [[int(coord) for coord in box.split(",")] for box in data]
```

```python
def connect_boxes(boxes: list[list[int]], part2: bool = False) -> int:
    distances = []

    for i, [x, y, z] in enumerate(boxes):
        for j, [x2, y2, z2] in enumerate(boxes[i + 1 :]):
            dist = (x2 - x) ** 2 + (y2 - y) ** 2 + (z2 - z) ** 2
            distances.append((dist, i, i + j + 1))

    distances.sort()

    graph = nx.Graph()

    for pair, (d, i, j) in enumerate(distances):
        graph.add_edge(i, j)

        if not part2 and pair + 1 == 1000:
            break
        if part2:
            for subgraph in nx.connected_components(graph):
                if len(subgraph) == len(boxes):
                    return boxes[i][0] * boxes[j][0]
                else:
                    break

    subgraphs = sorted(list(map(len, nx.connected_components(graph))), reverse=True)

    return subgraphs[0] * subgraphs[1] * subgraphs[2]
```

```python
connect_boxes(boxes)
```

```python
connect_boxes(boxes, True)
```
