```python
from pathlib import Path
import scipy
```

```python
data = Path("../Data/Day10.txt").read_text().splitlines()
```

```python
def get_machines(data: list[str]):

    machines = []

    for line in data:
        buttons = []
        target_lights = []
        joltage = []
        for s in line.split():
            
            if s.startswith("["):
                target_lights = [c == '#' for c in s[1:-1]]
            elif s.startswith("("):
                button = [int(c) for c in s[1:-1].split(",")]
                buttons.append(button)
            else:
                joltage = [int(c) for c in s[1:-1].split(",")]

        machines.append((target_lights, buttons, joltage))

    return machines

def set_joltage(buttons, joltage):
    a = [[0 for _ in range(len(buttons))] for __ in range(len(joltage))]

    for i in range(len(buttons)):
        for button in buttons[i]:
            a[button][i] = 1

    joltage_constraint = scipy.optimize.LinearConstraint(a, joltage, joltage)

    res = scipy.optimize.milp([1]*len(buttons), integrality=[1]*len(buttons), constraints = [joltage_constraint])

    if res.success:
        return res.fun
```

```python
machines = get_machines(data)
```

```python
sum(set_joltage(buttons, joltage) for _lights, buttons, joltage in machines)
```
