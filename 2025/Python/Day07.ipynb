```python
from pathlib import Path
```

```python
data = Path("../Data/Day07.txt").read_text().splitlines()
```

```python
def split_beam(data: list[str]) -> int:
    beams = {len(data[0]) // 2}
    splits = 0

    for row in data:
        beams2 = set()
        for beam in beams:
            

            if row[beam] == '^':
                splits += 1

                beams2.add(beam+1)
                beams2.add(beam-1)
            else:
                beams2.add(beam)
        
        beams = beams2
    
    return splits

def travel_beam(data: list[str], i: int, j: int, cache: dict[tuple[int, int], int]) -> int:
    if (i, j) in cache:
        return cache[i, j]
    if i+1 == len(data):
        return 1
    if data[i][j] != '^':
        cache[i, j] =  travel_beam(data, i+1, j, cache)
        return cache[i, j]
    else:
        left = travel_beam(data, i+1, j-1, cache)
        right = travel_beam(data, i+1, j+1, cache)

        cache[i, j] = left + right

        return cache[i, j]
```
