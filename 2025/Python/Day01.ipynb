```python
from pathlib import Path
```

```python
data = Path("../Data/Day01.txt").read_text().splitlines()
```

```python
def get_password(spins: list[str], part2: bool = False):

    dial = 50

    password = 0

    for spin in spins:
        spin_amount = int(spin[1:])

        if part2:
            password += spin_amount // 100

        spin_amount %= 100

        if spin[0] == "L":
            spin_amount *= -1

        if part2:
            if dial + spin_amount > 100:
                password += 1
            if dial != 0 and dial + spin_amount < 0:
                password += 1

        dial = dial + spin_amount

        dial %= 100

        if dial == 0:
            password += 1

    return password
```

```python
get_password(data)
```

```python
get_password(data, True)
```
