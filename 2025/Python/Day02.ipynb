```python
from pathlib import Path
```

```python
data = Path("../Data/Day02.txt").read_text().split(",")
data = [d.split("-") for d in data]
```

```python
def tile_num(num: int, n: int):
    mult = 10 ** len(str(num))

    total = num

    for _ in range(1, n):
        total *= mult
        total += num

    return total


def invalid_ids(data: list[list[str]], part2: bool = False):

    ids = set()

    for start, end in data:

        max_reps = len(end) if part2 else 2

        for reps in range(2, max_reps + 1):

            if len(start) % reps == 0:
                prefix = int(start[0 : len(start) // reps])

                num = tile_num(prefix, reps)

                while num <= int(end):
                    if num >= int(start):
                        ids.add(num)

                    prefix += 1
                    num = tile_num(prefix, reps)

            if len(end) > len(start) and len(end) % reps == 0:

                prefix = int(end[0 : len(end) // reps])

                num = tile_num(prefix, reps)

                while num <= int(end):
                    if num >= int(start):
                        ids.add(num)

                    prefix += 1
                    num = tile_num(prefix, reps)

    return sum(ids)
```

```python
invalid_ids(data)
```

```python
invalid_ids(data, True)
```
