```python
from pathlib import Path
from functools import reduce
```

```python
data = Path("../Data/Day03.txt").read_text().splitlines()
data = [[int(digit) for digit in line] for line in data ]
```

```python
def total_joltage(data: list[list[int]], flips: int):
    total = 0
    for batteries in data:
        best_voltage = [0] * flips

        for i, voltage in enumerate(batteries):
            for j in range(flips):
                if (
                    i >= j
                    and len(batteries) - i >= flips - j
                    and voltage > best_voltage[j]
                ):
                    best_voltage[j] = voltage
                    best_voltage[j + 1 :] = [0] * (flips - j - 1)
                    break

        total += reduce(lambda acc, e: 10 * acc + e, best_voltage, 0)

    return total
```

```python
total_joltage(data, 2)
```

```python
total_joltage(data, 12)
```
