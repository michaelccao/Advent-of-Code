```python
from pathlib import Path
import copy
```

```python
data = Path("../Data/Day04.txt").read_text().splitlines()
data = [[c for c in line] for line in data]
```

```python
def collect_rolls(data: list[str], part2: bool = False):

    grid = copy.deepcopy(data)
    total = 0

    while True:
        grid2 = copy.deepcopy(grid)
        subtotal = 0
        for i, line in enumerate(grid):
            for j, c in enumerate(line):
                if c != "@":
                    continue

                neighbors = 0

                for i2, j2 in [
                    (i - 1, j - 1),
                    (i - 1, j),
                    (i - 1, j + 1),
                    (i, j - 1),
                    (i, j + 1),
                    (i + 1, j - 1),
                    (i + 1, j),
                    (i + 1, j + 1),
                ]:
                    if (
                        0 <= i2 < len(grid)
                        and 0 <= j2 < len(line)
                        and grid[i2][j2] == "@"
                    ):
                        neighbors += 1

                if neighbors < 4:
                    grid2[i][j] = "."
                    subtotal += 1

        total += subtotal
        grid = grid2

        if subtotal == 0 or not part2:
            return total
```

```python
collect_rolls(data)
```

```python
collect_rolls(data, True)
```
