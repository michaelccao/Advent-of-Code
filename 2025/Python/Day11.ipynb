```python
from pathlib import Path
```

```python
data = Path("../Data/Day11.txt").read_text().splitlines()
```

```python
paths = {}

for line in data:
    start, ends = line.split(": ")
    
    for end in ends.split():
        if start in paths:
            paths[start].add(end)
        else:
            paths[start] = {end}
```

```python
def total_routes(paths, start, end, forbid, cache):
    if start in cache:
        return cache[start]
    
    if start == end:
        return 1
    
    if start == "out":
        return 0
    
    routes = 0

    for node in paths[start]:
        if node != forbid:
            routes += total_routes(paths, node, end, forbid, cache)

    cache[start] = routes
    
    return cache[start]
```

```python
total_routes(paths, "you", "out", "", {})
```

```python
svr_fft_dac_out = total_routes(paths, "svr", "fft", "dac", {}) * total_routes(paths, "fft", "dac", "", {}) * total_routes(paths, "dac", "out", "", {})
svr_dac_fft_out = total_routes(paths, "svr", "dac", "fft", {}) * total_routes(paths, "dac", "fft", "", {}) * total_routes(paths, "fft", "out", "", {})

part2 = svr_fft_dac_out + svr_dac_fft_out

print(part2)
```
