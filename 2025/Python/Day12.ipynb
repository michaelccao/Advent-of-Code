```python
from pathlib import Path
import numpy as np
```

```python
data = Path("../Data/Day12.txt").read_text().split("\n\n")
*presents, trees = data

presents = np.array([[[1 if c == '#' else 0 for c in row] for row in present.split("\n")[1:]] for present in presents])

trees = [tree.split(": ") for tree in trees.split("\n")]
trees = [([int(dim) for dim in tree[0].split("x")], np.array([int(ind) for ind in tree[1].split()])) for tree in trees]
```

```python
def can_tile(tree):
    # Specific to my input but generalizable

    # I can tile 4 present 2's
    # ######
    # ##..##
    # #....#
    # #....#
    # ##..##
    # ###### can fit, 1, 2, 3, 5 in middle

    # Similarly I can tile 4 present 5's
    # .####.
    # ##..##
    # #....#
    # #....#
    # ##..##
    # .####. can fit 1, 2, 3, 5 in middle

    [row, col], inv = tree

    tileable_presents = (row // 3) * (col // 3)

    presents_remaining = np.sum(inv) - tileable_presents

    squeeze = inv[2] // 4 + inv[5] // 4

    squeezable = inv[1] + inv[2] % 4 + inv[3] + inv[5] % 4

    presents_remaining -= min(squeeze, squeezable)

    return presents_remaining
```

```python
sum([can_tile(tree) <= 0 for tree in trees]) # Lucky guess, I suppose!
```

```python
# Turns out, you can be even cheekier, didn't even need to figure out packing
present_areas = np.sum(presents, axis = (1, 2))

sum(row*col - np.sum(present_areas*inv) >= 0 for [row, col], inv in trees)
```
