```python
from pathlib import Path
from functools import reduce
```

```python
data = Path("../Data/Day05.txt").read_text().splitlines()
```

```python
def get_fresh(data: list[str]) -> tuple[int, int]:

    total = 0

    fresh = []

    for line in data:
        if not line:
            continue

        if "-" in line:
            lower, upper = line.split("-")
            fresh.append((int(lower), int(upper)))
        else:
            ingredient = int(line)

            for lower, upper in fresh:
                if lower <= ingredient <= upper:
                    total += 1
                    break

    for i in range(len(fresh)):
        x0, x1 = fresh[i]
        for j in range(i + 1, len(fresh)):
            y0, y1 = fresh[j]

            if x0 <= y0 <= x1:
                if y1 > x1:
                    fresh[j] = (x1 + 1, y1)
                else:
                    fresh[j] = (1, 0)
            elif x0 <= y1 <= x1:
                if y0 < x0:
                    fresh[j] = (y0, x0 - 1)
                else:
                    fresh[j] = (1, 0)

    return total, reduce(lambda acc, e: acc + e[1] - e[0] + 1, fresh, 0)
```

```python
p1, p2 = get_fresh(data)

print(p1, p2, sep="\n")
```
