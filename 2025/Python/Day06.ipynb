```python
from pathlib import Path
from functools import reduce
import copy
```

```python
data = Path("../Data/Day06.txt").read_text().splitlines()
```

```python
def calculate(data: list[str]):
    data = copy.copy(data)
    ops = data.pop().split()
    nums = [[int(num) for num in row.split()] for row in data]

    total = 0

    for i, op in enumerate(ops):
        if op == "+":
            total += reduce(lambda acc, e: acc + e, [row[i] for row in nums], 0)
        elif op == "*":
            total += reduce(lambda acc, e: acc * e, [row[i] for row in nums], 1)

    return total


def calculate2(data: list[str]):
    data = copy.copy(data)
    ops = data.pop()

    current_op = "+"

    total = 0
    subtotal = 0
    for i, c in enumerate(ops):
        if c == "+":
            total += subtotal
            subtotal = 0
            current_op = "+"
        elif c == "*":
            total += subtotal
            subtotal = 1
            current_op = "*"

        try:
            num = int("".join(row[i] for row in data))

            if current_op == "+":
                subtotal += num
            elif current_op == "*":
                subtotal *= num
        except:
            pass

    return total + subtotal
```

```python
calculate(data)
```

```python
calculate2(data)
```
