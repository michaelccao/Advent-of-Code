```python
from pathlib import Path
import numpy as np
import re
from math import prod
from collections import defaultdict
```

```python
data = Path('../Data/Day14.txt').read_text()
```

```python
patt = r"p=(-?\d+),(-?\d+) v=(-?\d+),(-?\d+)"
```

```python
robots = [[int(num) for num in robot] for robot in re.findall(patt, data)]
```

```python
height = 103
width = 101
```

```python
robots2 = [[(robot[0] + robot[2]*100) % width, (robot[1] + robot[3]*100) % height] for robot in robots]
```

```python
quadrants = [(0, height//2, 0, width//2),
             (0, height//2, width//2+1, width),
             (height//2+1, height, 0, width//2),
             (height//2+1, height, width//2+1, width)]

total = 0


prod([len([(x,y) for x, y in robots2 if xl <= x < xu and yl <= y < yu]) for yl, yu, xl, xu in quadrants])
```

```python
patterns = defaultdict(list)

# After looking at a lot of patterns, notice a more ordered patterns appears at 71 and every cycle of 101 (i.e. x coordinate cycles back)
for t in range(71, height*width+101, 101):

    grid = np.zeros((height, width), dtype=str)
    grid[:, :] = '_'

    for x, y, dx, dy in robots:
        x2 = x + dx*t
        y2 = y + dy*t

        x2 %= width
        y2 %= height

        grid[y2, x2] = '#'
    
    pattern = '\n'.join(''.join(line) for line in grid)

    patterns[pattern].append(t)
```
