## Don't wanna be your monkey wrench

```python
from pathlib import Path
import numpy as np
```

```python
monkeys = Path('monkeys.txt').read_text().split('\n')
```

```python
monkey_items = {}
monkey_op = {}
monkey_test = {}

current_monkey = 0

for line in monkeys:
    if line.startswith('Monkey'):
        current_monkey = int(line.split(' ')[-1][0:-1])
        
    if line.startswith('  Starting items'):
        monkey_items[current_monkey] = [int(item) for item in ''.join(line.split(' ')[4:]).split(',')]
    
    elif line.startswith('  Operation'):
        op, num = line.split(' ')[-2:]
        
        monkey_op[current_monkey] = [op, num]
        
    elif line.startswith('  Test'):
        div = int(line.split(' ')[-1])
        
        monkey_test[current_monkey] = [div]
    
    elif line.startswith('    If true'):
        target = int(line.split(' ')[-1])
        
        monkey_test[current_monkey] += [target]
    
    elif line.startswith('    If false'):
        target = int(line.split(' ')[-1])
        
        monkey_test[current_monkey] += [target]
        

        
num_monkeys = len(monkey_items)
num_rounds = 20

monkey_inspects = [0,]*num_monkeys

for i in range(num_rounds):
    for j in range(num_monkeys):
        num_items = len(monkey_items[j])
        for k in range(num_items):
            item = monkey_items[j][0]
            
            op, num = monkey_op[j]
            
            if num != 'old':
                num = int(num)
                
                if op == '+':
                    item += num
                elif op == '*':
                    item *= num
            else:
                if op == '+':
                    item += item
                elif op == '*':
                    item *= item
                
            monkey_inspects[j] += 1
            
            item //= 3
            
            monkey_items[j].pop(0)
            if item % monkey_test[j][0] == 0:
                target = monkey_test[j][1]
            else:
                target = monkey_test[j][2]
            
            monkey_items[target].append(item)
            
monkey_inspects.sort()

print(monkey_inspects[-1]*monkey_inspects[-2])
```
